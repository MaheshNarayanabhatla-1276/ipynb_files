```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
```

```python
driver = webdriver.Chrome('C:\\Users\\numam\\Downloads\\chromedriver')
```

```python
driver.get('https://www.amazon.jobs/en-gb/job_categories/software-development')
```

```python
from time import sleep
```

```python
career=[]
career0=[]
for j in range(1,729):
    print(j)
    JT=driver.find_elements(By.CLASS_NAME,"job-title")
    for k in range(len(JT)):
        sleep(1.5)
        JT=driver.find_elements(By.CLASS_NAME,"job-title")
        JT.__getitem__(k).click()
        print(len(JT),k)
        career=[]
        job_title=driver.find_elements(By.CLASS_NAME,"title")
        career.append(job_title.__getitem__(0).text)

        # ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~    lOCATION    ~~~~~~~~~~~~~~~~~~~~~~~~~~~~

        job_location=driver.find_elements(By.CLASS_NAME,"association-content")
        career.append(job_location.__getitem__(0).text)
        Job_description=driver.find_elements(By.CLASS_NAME,"content")

        # ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~     QUALIFICATION          ~~~~~~~~~~~~~~~~~~~~~~

        qualification=[]
        eligibility=[]
        experience=[]
        for i in range(len(Job_description.__getitem__(1).text.split("\n"))):
            if Job_description.__getitem__(1).text.split("\n")[i].find('degree ')>=1 or Job_description.__getitem__(1).text.split("\n")[i].find('achelor')>=1 :
                qualification.append(Job_description.__getitem__(1).text.split("\n")[i])
            if Job_description.__getitem__(1).text.split("\n")[i].find("ears")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("onth")>=1:
                eligibility.append(Job_description.__getitem__(1).text.split("\n")[i])
            if Job_description.__getitem__(1).text.split("\n")[i].find("xper")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("nowle")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("be able to")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("Basic")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("esponsi")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("bility")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("to have")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("ust H")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("ust-H")>=1 or Job_description.__getitem__(1).text.split("\n")[i].find("to-H")>=1:
                experience.append(Job_description.__getitem__(1).text.split("\n")[i])


        career.append(qualification)
        print("qualification done")
        career.append(experience)
        print("experience done")
        career.append(eligibility)
        print("eligibility done")




        #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~  JOB DESCRIPTION ~~~~~~~~~~~~~~~
        career.append(Job_description.__getitem__(1).text.split("\n"))
        """
        role \ title
        location
        qualification
        eligibility
        experience
        job description

        """
        
        career0.append(career)
        career=[]
        driver.back()
    sleep(0.5)
    driver.find_element(By.CLASS_NAME,"btn.circle.right").click()
    print(j)
```
