```python
an_array = np.array(penguins_df["body_mass_g"])
mean = np.mean(an_array)
standard_deviation = np.std(an_array)
distance_from_mean = abs(an_array - mean)
max_deviations = .25
not_outlier = distance_from_mean < max_deviations * standard_deviation
no_outliers = an_array[not_outlier]
```

```python
pdf=stats.norm.pdf(np.sort(no_outliers),np.mean(no_outliers),np.std(no_outliers))
```

```python
from sklearn.cluster import KMeans
x=((penguins_df["bill_length_mm"]).dropna()).values
y=((penguins_df["bill_depth_mm"]).dropna()).values
t=[]
for i in range(len(x)):
    t.append([x[i],y[i]])
kmeans=KMeans(n_clusters=3).fit(t)
```

```python
kmeans.labels_
```

```python
clusters=kmeans.cluster_centers_
```

```python
clusters[0]
```

```python
plt.scatter(penguins_df["bill_length_mm"],penguins_df["bill_depth_mm"])
val0=[]
val1=[]
for j in range(len(clusters)):
    val0.append(clusters[j][0])
    val1.append(clusters[j][1])
plt.scatter(val0,val1)
```
