```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome('C:\\Users\\numam\\Downloads\\chromedriver')
```

```python
driver.get('https://ibegin.tcs.com/iBegin/jobs/search')
```

```python
from time import sleep
```

```python
desc=driver.find_element(By.CLASS_NAME,"ng-binding[data-ng-bind='title( j.title )']")
desc.text
click_=driver.find_element(By.LINK_TEXT,desc.text)
click_.click()
career=[]
career0=[]
for iterations in range(2882):
    #---------------------- iteration show ------------------
    print(iterations)
    
    
    # Loading time
    sleep(4)
    # --------------------    next page    -------------------------
    
    img=driver.find_element(By.CLASS_NAME,"hand[data-ng-click='arrowRight()']")
    sleep(0.5)
    img.click()
    
    #  Loading time 
    sleep(1.5)
    
    #  ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~  TITLE  ~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
    job_title=driver.find_elements(By.CLASS_NAME,"ng-binding[data-ng-bind='jobDescription.title']")
    career.append(job_title.__getitem__(0).text)
    
    # ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~    lOCATION    ~~~~~~~~~~~~~~~~~~~~~~~~~~~~
    
    job_location=driver.find_elements(By.CLASS_NAME,"ng-binding[data-ng-bind='jobDescription.location']")
    career.append(job_location.__getitem__(0).text)
    Job_description=driver.find_element(By.CLASS_NAME,"job-description")
    
    # ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~     Eucation          ~~~~~~~~~~~~~~~~~~~~~~
    
    qualification=[]
    for i in range(len(Job_description.text.split("\n"))):
        if Job_description.text.split("\n")[i].startswith('Qualifications'):
            qualification.append(Job_description.text.split("\n")[i])
    career.append(qualification)
    
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~   skills  ~~~~~~~~~~~~~~~~~~~~~~~~~~
    skill=[]
    for i in range(len(Job_description.text.split("\n"))):
        if Job_description.text.split("\n")[i].find("ears")>=1 or Job_description.text.split("\n")[i].find("onth")>=1 or Job_description.text.split("\n")[i].find("to have")>=1 or Job_description.text.split("\n")[i].find("ust H")>=1 or Job_description.text.split("\n")[i].find("ust-H")>=1 or Job_description.text.split("\n")[i].find("to-H")>=1:
            skill.append(Job_description.text.split("\n")[i])
    career.append(skill)
    
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~   EXPERIENCE  ~~~~~~~~~~~~~~~~~~~~~~~~~~
    experience=[]
    for i in range(len(Job_description.text.split("\n"))):
        if Job_description.text.split("\n")[i].find("xper")>=1 or Job_description.text.split("\n")[i].find("nowle")>=1 or Job_description.text.split("\n")[i].find("be able to")>=1 or Job_description.text.split("\n")[i].find("Basic")>=1 or Job_description.text.split("\n")[i].find("esponsi")>=1 or Job_description.text.split("\n")[i].find("bility")>=1:
            experience.append(Job_description.text.split("\n")[i])
    career.append(experience)
    
    
    
    
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~  JOB DESCRIPTION ~~~~~~~~~~~~~~~
    career.append(Job_description.text.split("\n"))
    career0.append(career)
    career=[]
```
